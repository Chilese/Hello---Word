```python
#Importar a biblioteca, declarar o data frame que será usado.
#Duas funções são usadas.

import seaborn as sns
tips = sns.load_dataset("tips")
tips.head() 
```

#Biblioteca de dados

Data frame da prória  biblioteca Seaborn, utilizado para análises iniciais. 

**Total_bill:** Custo total da refeição em dólares.

**Tip:** Valor da gorjeta em dólares.

**Sex:** Sexo (Feminino / Masculino).

**Smoker:** Se a área escolhida era para fumantes ou não.

**Day:** Dia da semana.

**Time:** Almoço (Lunch) ou jantar (Dinner).

**Size:** Tamanho da refeição.

# Relplot

*Interface de nível de figura para desenhar gráficos relacionais em um FacetGrid.*

Essa função fornece acesso a várias funções de nível de eixo diferentes que mostram o relacionamento entre duas variáveis ​​com mapeamentos semânticos de subconjuntos. 

```python
# relplot permite uma análise simples da relação entre duas variáveis em um data frame.
# Como ser verifica, quanto maior o valor da refeição, maior o valor da gorjeta. 
sns.relplot(x="total_bill", y="tip", data= tips)
```

```python
# O argumento hue permite detalhar os valores de uma determinada variável. 
# Usei como parametro se fuma ou não, mas poderia ter sido qualquer outra variável.

sns.relplot(x= "total_bill", y="tip", hue= "smoker", data= tips)
```

```python
# É possível ainda definir uma forma para cada um dos valores, um style. 

sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=tips)
```

```python
# Poderíamos inserir mais uma variável:
#Usando hue como um marcador e style como outro. 

sns.relplot(x="total_bill", y="tip", hue="smoker", style="sex", data=tips)
```

```python
# Até aqui os marcadores foram usados em dados quase que boleanos "sim ou não" e "Male ou Female"
# Poderiamos usar o marcador para dados quantitativos. Por exemplo, "size". 

sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

```python
# Mas também existe a opção size que poderia ser usada para expressar essa informação de forma ainda mais intuitiva. 

sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
# A opção sizes permite ampliar e criar uma escala para os valores plotados.
# Ficou evidente quais foram as maiores refeições. 

sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,300), data=tips)
```

# Scatterplot

*Desenhe um gráfico de dispersão com possibilidade de vários agrupamentos semânticos.*

A relação entre xe ypodem ser mostrados para diferentes subconjuntos dos dados utilizando as hue, sizee styleparâmetros. Esses parâmetros controlam qual semântica visual é usada para identificar os diferentes subconjuntos. É possível mostrar até três dimensões independentemente usando todos os três tipos semânticos, mas esse estilo de trama pode ser difícil de interpretar e geralmente ineficaz. Usar semântica redundante (ou seja, ambos huee stylepara a mesma variável) pode ser útil para tornar os gráficos mais acessíveis.

```python
# Dispersando os dados em quatro variáveis, marcando em duas variáveis.

sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

```python
import matplotlib.pyplot as plt
# Usando matplotlib para redimensionar a figura. 
# Criar um novo colab para estudar apenas o matplotlib. 

f, ax = plt.subplots(figsize=(15, 7))
sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips)
```

```python
# Ainda é possível usar paletas de cores para os dados (sequencial, divergente, quantitativos)

sns.relplot(x="total_bill", y="tip", hue="size", palette="Blues", size="size", sizes=(12,250), data=tips);
```
